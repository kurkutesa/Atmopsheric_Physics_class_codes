```python
#!/usr/bin/python 
# imports
import pylab

from scipy import ndimage

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab


# Add a couple of user defined functions
sys.path.append("/Users/scavallo/scripts/python_scripts")
import weather_modules as wm
import utilities_modules as um
from mstats import *

import warnings
warnings.filterwarnings("ignore")
```

```python
# User options
label_fontsize = 16
example_opt = 2
B_opt = 2 # B_opt = 1: Background error covariances are random 
          # B_opt = 2: Background error covariances are isotropic, centered on low pressure center
	      # B_opt = 3: Background error covariances are anisotropic, defined by a function
mean_obs_error = 1 # default is 1
mean_bg_error = 1 # default is 1

# Set for B_opt == 1
nens = 100 # Number of ensemble members
mean_background_error, std_background_error = 10,0.2
mean_background, std_background = 1020, 10
mean_obs, std_obs = 1000, 10 
std_obs_error = 0.2

# Set for B_opt == 2 or B_opt == 3
slp_mean_bg, slp_mean_obs = 1020, 1010 # Mean SLP from background (prior) and observations
slp_range = 1020-980 # Define a range of SLP
```

```python
if example_opt == 1:

   xb = mean_background + std_background*pylab.randn(nens);
   yo = mean_obs + std_obs*pylab.randn(nens);
   be = mean_background_error + std_background_error*pylab.randn(nens);
   oe = mean_obs_error + std_obs_error*pylab.randn(nens)

   bins = np.arange(980,1050,1)


   B = (be*np.transpose(be))/(nens-1)
   R = (oe*np.transpose(oe))/(nens-1)
   K = B*(B + R)**(-1)

   xa = xb + K*(yo - xb)
   mean_analysis = np.mean(xa)
   std_analysis = np.std(xa)

   fig = plt.figure(1)
   ax1 = fig.add_subplot(111)
   p1 = ax1.plot(bins,mlab.normpdf(bins,mean_background,std_background),'b',linewidth=2)
   p2 = ax1.plot(bins,mlab.normpdf(bins,mean_obs,std_obs),'r',linewidth=2)
   p3 = ax1.plot(bins,mlab.normpdf(bins,mean_analysis,std_analysis),'g',linewidth=2)
   ax1.grid(True)
   ax1.set_ylabel('Frequency',fontsize=label_fontsize)
   ax1.set_xlabel('Sea level pressure (hPa)',fontsize=label_fontsize)
   leg = ax1.legend(('Background (Prior)','Observations','Analysis (Posterior)'),'upper right', shadow=True)   
   figname = "da_pdf_posterior"
   save_name = figname + ".png"
   plt.savefig(save_name, bbox_inches='tight')
   
   plt.show()
else: 
   nxpts,nypts = 100, 100      
      
   x = np.arange(-nxpts,nxpts+0.1,1)
   y = np.arange(-nypts,nypts+0.1,1)   
   [X,Y] = np.meshgrid(x,y)
   Z = np.sqrt(X**2 + Y**2) 
   nx,ny = np.shape(Z)
   W = Z/np.max(Z)
     
   xshift = nxpts/4
   yshift = nypts/4
   
   # Define the region for an anomaly
   Z_bg = (np.sqrt( (X-xshift)**2 + (Y-yshift)**2) ) - nx     
   # Make the anomaly amplitude reasonable
   xfac = np.abs(slp_range/np.min(Z_bg))
   Z_anom = Z_bg*xfac
   # The background field equals the mean value plus the anomaly
   Z_bg = Z_anom + slp_mean_bg   
   
   Z_obs = (np.sqrt( (X+xshift)**2 + (Y+yshift)**2) ) - nx  
   # Make the anomaly amplitude reasonable   
   xfac = np.abs(slp_range/np.min(Z_obs))   
   Z_obs = Z_obs*xfac
   Z_obs = Z_obs + slp_mean_obs 
   
   H = np.ones_like(X).astype('f')
   
   if B_opt == 1:      
      be = mean_background_error + std_background_error*pylab.randn(ny,nx)                       
   elif B_opt == 2:
      xfac =  2*np.abs(mean_bg_error/np.min(Z_anom))
      if ( mean_bg_error < 0 ):
          xfac = -1*xfac       
      Z_anom = Z_anom*xfac
      be = Z_anom
   else:    
      xfac = (2*np.pi)/nx       
      x1 = x*xfac
      y1 = y*xfac
      

      #ZZ = ((X+20)/(2) )**2 + ((Y+20)/(1) )**2
      ZZ = np.arcsinh(X) + np.arcsinh(Y)   
      #ZZ = np.sinh(Y/nx) + np.cos(Y/nx)
        
      minmaxzz = np.min(ZZ),np.max(ZZ)
      minmaxzz = np.abs(minmaxzz)
      zfac = mean_bg_error/np.max(minmaxzz)
      ZZ = ZZ * zfac  
        
      #x2 = 0.05*np.sinh(x1)*np.cos(x1)             
      x2 = np.cos(2*x1) 
      #x2 = np.sin(x1)
      #y2 = x2
            
      
      [XX,YY] = np.meshgrid(x2,y)     
      be = np.zeros_like(X).astype('f')
      jind = 0     
      for jj in y:
         iind = 0
         for ii in x:	    
	    be[jind,iind] = XX[jind,iind]
	    iind += 1
	 jind += 1      
   
      #be = 2*ZZ*mean_bg_error
      be = 2*be*mean_bg_error   
   
   B = np.cov(be)
          
   oe = np.zeros_like(X).astype('f')
   aa = X/np.max(X)
   bb = Y/np.max(Y)
   oe[:,:] = (aa**2+bb**2)*mean_obs_error
   
   R = np.cov(oe)
  
   K = B*np.transpose(H)*( (R + (H*B*np.transpose(H)) )**(-1))        
   Z_analysis = Z_bg + K*(Z_obs - H*Z_bg)

   Z_inc = Z_analysis - Z_bg

   slps = [slp_mean_obs,slp_mean_bg]

   cfint = 4
   tmp = np.append(slp_mean_bg, slp_mean_obs)
   cbar_max = np.max(tmp)
   cbar_min = cbar_max - (30*cfint)

   minz = np.floor(np.min(Z_analysis))
   minz = int(minz)   
   for num in range(1, cfint+1):      
      if ( minz%cfint == 0 ):
         break
      else:
         minz = minz + 1
   maxz = np.ceil(np.max(Z_analysis))
   maxz = int(maxz)   
   for num in range(1, cfint+1):      
      if ( maxz%cfint == 0 ):
         break
      else:
         maxz = maxz - 1   
   
   tmp2 = np.append(mean_obs_error,mean_bg_error)
   tmp2max = 2*np.max(np.abs(tmp2))
   cflevs_err = np.arange(-tmp2max,tmp2max+0.1,0.2)                         
                            
   incguess =  np.abs(slp_mean_bg - slp_mean_obs)
   
                            
   cflevs = np.arange(minz-cfint,maxz+1,cfint)   
   cflevs2 = np.arange(cbar_min,cbar_max+1,cfint)   
   cflevs3 = np.arange(-incguess,incguess,1)
   
                            
   titletext1 = ('Background (solid), and Observations (dashed) SLP')   
   titletext2 = ('SLP Background error')
   titletext3 = ('SLP Observation error') 
   titletext4 = ('Analysis increment') 
   titletext5 = ('Analysis (colors), background (solid), and Obs. (dashed) SLP')   

   fig = plt.figure(1)
   ax1 = fig.add_subplot(111)    
   cf2 = plt.contour(X, Y, Z_bg, cflevs2,
                           colors = 'k',
                           hold='on')
   for c in cf2.collections:
       c.set_linestyle('solid')   

   plt.clabel(cf2, inline=1, fontsize=10, fmt='%i')      
   cf3 = plt.contour(X, Y, Z_obs, cflevs2,
                           colors = 'k',
                           hold='on')
   for c in cf3.collections:
       c.set_linestyle('dashed')   
   plt.clabel(cf3, inline=1, fontsize=10, fmt='%i')   
   ax1.set_title(titletext1)

   fig = plt.figure(2)
   ax1 = fig.add_subplot(111) 
   cf1 = plt.contourf(X,Y,be,cmap=plt.cm.bwr,levels=cflevs_err)
   cbar = plt.colorbar()   
   ax1.set_title(titletext2)

   fig = plt.figure(3)
   ax1 = fig.add_subplot(111) 
   cf1 = plt.contourf(X,Y,oe,cmap=plt.cm.bwr,levels=cflevs_err)
   cbar = plt.colorbar()   
   ax1.set_title(titletext3)
   

   plt.show()
```
